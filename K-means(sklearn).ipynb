```python
%matplotlib inline

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (12, 7)
plt.style.use('ggplot')
```

```python
data = pd.read_csv('/Users/yuyizhang/Documents/Algorithm learning/cluster_example.csv')
```

```python
f1 = data['V1'].values
f2 = data['V2'].values

X = np.array(list(zip(f1, f2)))
plt.scatter(f1, f2, c='black', s=6)
```

```python
from sklearn.cluster import KMeans
```

```python
# Fit a K-Means model to the data
kmeans_3 = KMeans(n_clusters=3)
kmeans_iris = kmeans_3.fit(data[['V1', 'V2']])
```

```python
y_kmeans = kmeans_3.predict(data)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=8,cmap='viridis')
# rainbow; viridis; Set1
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=8, cmap='viridis')
centers = kmeans_3.cluster_centers_
plt.scatter(centers[:,0], centers[:, 1], c='black', s=100, marker='x')
```

```python
inertia = []
for k in range(1, 8):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(X)
    inertia.append(np.sqrt(kmeans.inertia_))
```

```python
plt.plot(range(1, 8), inertia, marker='s');
plt.xlabel('$k$')
plt.ylabel('$J(C_k)$');
```
